```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def process_df(file_name):
    df=pd.read_csv(file_name,names=["Time:s","Time:us",'Window'],dtype=float)
    df["Time"]=df["Time:s"]*1000+df["Time:us"]*(1.0/1000) 
    return df
```

# 10% downlink packet Loss

### Congestion Window

```python
df=process_df("CWND_downlink.csv")
```

```python
df
```

```python
plt.plot(df["Time"],df["Window"],)
```

#### Zooming in

```python
plt.plot(df[(df.index>1000) & (df.index<1100)]["Time"],df[(df.index>1000) & (df.index<1100)]["Window"])
```

### Timeout

```python
df=process_df("Timeout_downlink.csv")
```

```python
plt.plot(df["Time"],df["Window"])
```

#### Zooming in

```python
plt.plot(df["Time"],df["Window"])
plt.ylim(0,100)
```
